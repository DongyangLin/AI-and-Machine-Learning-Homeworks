### Problem 1
Write a Python program to convert a 10x2 Cartesian coordinate matrix to polar coordinates by using NumPy.

```python
import numpy as np
# create a 10x2 Cartesian coordinate matrix
a=np.array([[0,0],[0,1],[1,0],[1,1],[2,0],[0,2],[2,1],[1,2],[2,2],[3,3]])
print("Cartesian coordinate matrix:")
print(a)
print()
# convert the Cartesian coordinate matrix to polar coordinate matrix
b=np.zeros((10,2))
for i in range(10):
    b[i][0]=np.sqrt(a[i][0]**2+a[i][1]**2) # calculate the radius
    # calculate the angle
    if a[i][0]==0:
        if a[i][1]==0: # a special case: (0,0)
            b[i][1]=0
            continue
        else: # when x=0,y!=0
            b[i][1]=90
            continue
    b[i][1]=np.degrees(np.arctan(a[i][1]/a[i][0])) 
print("Polar coordinate matrix:")
print(b)
```

### Problem 2
Using Numpy, read all 150 data from iris.csv about iris’s Sepal.Length, Sepal.Width, Petal.Length, and Petal.Width to an ndarray. Then calculate their mean, median and standard deviation. Also, draw a scatter plot of Sepal.Length and Sepal.Width as well as a scatter plot of Petal.Length and Petal.Width.

```python
# load the data into a ndarray
iris=np.loadtxt('iris.csv',delimiter=',',skiprows=1,usecols=(0,1,2,3))
# calculate the means
slenmean=np.mean(iris[:,0])
swidmean=np.mean(iris[:,1])
plenmean=np.mean(iris[:,2])
pwidmean=np.mean(iris[:,3])
print(f"mean of Sepal.Length is {slenmean}")
print(f"mean of Sepal.Width is {swidmean}")
print(f"mean of Petal.Length is {plenmean}")
print(f"mean of Petal.Width is {pwidmean}")
# calculate the medians
slenmedian=np.median(iris[:,0])
swidmedian=np.median(iris[:,1])
plenmedian=np.median(iris[:,2])
pwidmedian=np.median(iris[:,3])
print(f"median of Sepal.Length is {slenmedian}")
print(f"median of Sepal.Width is {swidmedian}")
print(f"median of Petal.Length is {plenmedian}")
print(f"median of Petal.Width is {pwidmedian}")
# calculate the standard deviations
slenstd=np.std(iris[:,0])
swidstd=np.std(iris[:,1])
plenstd=np.std(iris[:,2])
pwidstd=np.std(iris[:,3])
print(f"standard deviation of Sepal.Length is {slenstd}")
print(f"standard deviation of Sepal.Width is {swidstd}")
print(f"standard deviation of Petal.Length is {plenstd}")
print(f"standard deviation of Petal.Width is {pwidstd}")
import matplotlib.pyplot as plt
# plot figures
x1=iris[:,0]
y1=iris[:,1]
plt.scatter(x1,y1)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Figure 1')
plt.grid(True)
plt.show()
x2=iris[:,2]
y2=iris[:,3]
plt.scatter(x2,y2,c='r')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Figure 2')
plt.grid(True)
plt.show()
```
