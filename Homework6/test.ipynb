```python
import numpy as np
import matplotlib.pyplot as plt

# 设置随机种子，以确保生成相同的数据
np.random.seed(0)

# 生成随机数据
X = np.random.rand(100, 1)
y = 2 * X + 1 + 0.1 * np.random.randn(100, 1)

print(X)

# 绘制数据点
plt.scatter(X, y, marker='o', label='Data Points')

# 添加标签和标题
plt.xlabel('X')
plt.ylabel('y')
plt.title('Scatter Plot of Random Data Points')

# 显示图例
plt.legend()

# 显示图形
plt.show()
```
